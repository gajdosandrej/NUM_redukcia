 # <font color=brown> Vety o strednej hodnote, Taylorove vety</font>

Pod označením $C^n\langle a,b\rangle$ budeme rozumieť triedu reálných funkcií, ktoré sú definované na intervale $\langle a,b\rangle$ a sú na tomto intervale $n$-krát spojite diferencovateľné (majú spojité derivácie až do rádu $n$). Označenie $C\langle a,b\rangle$ alebo tiež niekedy $C^0\langle a,b\rangle$ bude reprezentovať triedu funkcií spojitých na intervale $\langle a,b\rangle$. Ďalej označenie $C^1\langle a,b\rangle$ bude reprezentovať tie funkcie, ktoré sú spojité na $\langle a,b\rangle$ a tiež ich prvé derivácie sú spojité na $\langle a,b\rangle$. Podobne $f\in C^2\langle a,b\rangle$, $f$ je spojitá na $\langle a,b\rangle$ a taktiež jej prvá i druhá derivácia sú spojité na $\langle a,b\rangle$, atď. 

--- 

### <font color=green> Veta 1 (Prvá veta o strednej hodnote) </font> 

Nech $f\in C\langle a,b\rangle$ a nech $k$ je ľubovoľné reálne číslo medzi $m=\underset{a\le x\le b}{\mathrm{min}}\,f(x)$ a $M=\underset{a\le x\le b}{\mathrm{max}}\,f(x)$. Potom existuje reálne číslo $c\in\langle a,b\rangle$ také, že platí $f(c)=k$. 

---

Vetu o strednej hodnote výstižne ilustruje nasledujúci obrázok. 

 

$$\text{Zdroj: Ackleh et al. (2009).}$$

---

#### Poznámka 1 

* V angličtine sa Veta 1 zvykne označovať ako *Intermediate value theorem*. Ide o akúsi obdobu resp. modifikáciu tzv. Darbouxovej vety (viď úvodný kurz matematickej analýzy). 


* Niekedy sa zvykne Veta 1 uvádzať v tvare: Nech $f\in C\langle a,b\rangle$ a nech $k$ leží medzi hodnotami $f(a)$, $f(b)$, pričom $f(a)\neq f(b)$; potom existuje $c\in(a,b)$ také, že $f(c)=k$.   


* Iný pohľad na vetu o strednej hodnote vlastne hovorí, že funkcia $f$ spojitá na intervale $\langle a,b \rangle$ nadobúda na tomto intervale všetky hodnoty medzi $f(a)$, $f(b)$. 

---

--- 

### <font color=green> Veta 2 (O strednej hodnote pre integrály) </font> 

Nech $f$ je spojitá funkcia na intervale $\langle a,b\rangle$, $w$ je Riemannovsky integrovateľná funkcia na intervale $\langle a,b\rangle$ a predpokladajme, že $w(x)\ge0$ (resp. $w(x)\le0$) pre $x\in\langle a,b\rangle$. Potom existuje bod $c\in\langle a,b\rangle$ taký, že platí 

$$\displaystyle\int_{a}^{b}w(x)f(x)\,\mathrm{d}x=f(c)\displaystyle\int_{a}^{b}w(x)\,\mathrm{d}x.$$

---

---

#### Poznámka 2 

* Funkcia $w(x)$ vystupujúca vo Vete 1, bude v rámci numerickej analýzy zväčša spojitá ale nemusí tomu tak byť vždy!  


* Niekedy sa uvádza modifikácia Vety 2 v tom zmysle, že $c\in(a,b)$. 

---

Ak zvolíme vo Vete 2 špeciálne $w(x)=1$, pričom uvažujeme $f(x)=8-2x$, tak nasledujúci obrázok názorne ilustruje o čom táto veta pojednáva ...

### <font color=green> Veta 3 (Taylorova) </font> 

Nech $f\in C^{n+1}\langle a,b\rangle$ a $x_0\in \langle a,b\rangle$. Potom pre ľubovoľné $x\in \langle a,b\rangle$ platí 

$$f(x)=T_n(x)+R_{n}(x),$$

kde $T_n(x)=f(x_0)+f'(x_0)(x-x_0)+\ldots+\dfrac{f^{(n)}(x_0)(x-x_0)^n}{n!}=\displaystyle\sum_{k=0}^{n}\dfrac{1}{k!}f^{(k)}(x_0)(x-x_0)^k$ a 

$$R_n(x)=\dfrac{1}{n!}\displaystyle\int_{x_0}^{x}f^{(n+1)}(t)(x-t)^n\,\mathrm{d}t \text{ (integrálny tvar zvyšku)}.$$

Naviac existuje $\xi=\xi(x)$ medzi $x_0$ a $x$ také, že platí 

$$R_n(x)=\dfrac{f^{(n+1)}(\xi(x))(x-x_0)^{n+1}}{(n+1)!} \text{ (Lagrangeov tvar zvyšku)}.$$

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
from sympy.utilities.lambdify import lambdify

# aproximacia 1-sin(x) 
def multiple_formatter(denominator=2, number=np.pi, latex='\pi'):
    def gcd(a, b):
        while b:
            a, b = b, a%b
        return a
    def _multiple_formatter(x, pos):
        den = denominator
        num = np.int(np.rint(den*x/number))
        com = gcd(num,den)
        (num,den) = (int(num/com),int(den/com))
        if den==1:
            if num==0:
                return r'$0$'
            if num==1:
                return r'$%s$'%latex
            elif num==-1:
                return r'$-%s$'%latex
            else:
                return r'$%s%s$'%(num,latex)
        else:
            if num==1:
                return r'$\frac{%s}{%s}$'%(latex,den)
            elif num==-1:
                return r'$\frac{-%s}{%s}$'%(latex,den)
            else:
                return r'$\frac{%s%s}{%s}$'%(num,latex,den)
    return _multiple_formatter

class Multiple:
    def __init__(self, denominator=2, number=np.pi, latex='\pi'):
        self.denominator = denominator
        self.number = number
        self.latex = latex

    def locator(self):
        return plt.MultipleLocator(self.number / self.denominator)

    def formatter(self):
        return plt.FuncFormatter(multiple_formatter(self.denominator, self.number, self.latex))

def interactive_sin_plot(N=3):
    
    xx = np.linspace(0, 2*np.pi)
    
    ts = lambdify(x, taylor_sin(N).removeO(), 'numpy')
    
    fig, ax = plt.subplots()
    ax.plot(xx, np.sin(xx))
    ax.plot(xx, ts(xx))
    ax.set_ylim([-1.5, 1.5])
    ax.set_ylabel(r'$\sin(x)$')
    ax.set_xlabel(r'$x$')
    plt.grid()
    
    ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
    ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 12))
    ax.xaxis.set_major_formatter(plt.FuncFormatter(multiple_formatter()))
    
    plt.show()
    plt.close(fig)
    
interactive_sin_plot = interactive(interactive_sin_plot, N=IntSlider(min=3, max=20, description='No. of Terms:'))
interactive_sin_plot
```

#### <font color=blue> Príklad 1 </font> 

Gaussova funkcia centrovaná v bode $x=0$ je daná predpisom 

\begin{equation}
    f(x) = \exp\left( -\frac{x^2}{b} \right). \nonumber
\end{equation}

Derivácie:

\begin{align*}
    f^\prime(x)         &= -\frac{2x}{b}\exp\left( \frac{-x^2}{b}\right) \\
    f^{\prime\prime}(x) &= \left( \frac{4x^2-2b}{b^2} \right) \exp\left( -\frac{x^2}{b}\right) \\
    f^{\prime\prime\prime}(x) &= \frac{4x ( 3b-2x^2 )}{b^3} \exp\left( -\frac{x^2}{b}\right) 
\end{align*} 

Aproximácie uvažovanej funkcie $f$ Taylorovym polynómom obsahujúce rôzne počty členov sú znázornené pomocou nasledujúceho kódu: 


```python
%matplotlib inline

def f(x,b):
    return np.exp(-x**2/b)
def df(x,b):
    return -2*x/b * f(x,b)
def d2f(x,b):
    return (4*x**2-2*b)/b**2 * f(x,b)
def d3f(x,b):
    return 4*x*(3*b-2*x**2)/b**3 * f(x,b)

b = 1
xhi = 3
x = np.linspace(-xhi,xhi,1000)
x0 = -.2

# Taylorov rozvoj v okoli bodu x0
term1 = f(x0,b)*np.ones_like(x)
term2 = df(x0,b) * (x-x0)
term3 = 0.5 * d2f(x0,b) * (x-x0)**2
term4 = 1/6 * d3f(x0,b) * (x-x0)**3

plt.plot( x0,f(x0,b), 'ko' )
plt.plot( x, f(x,b), 'k-', label='f(x)' )
plt.plot( x, term1, '--', label='1 člen')
plt.plot( x, term1+term2, '--', label='2 členy')
plt.plot( x, term1+term2+term3, '--', label='3 členy')
plt.plot( x, term1+term2+term3+term4, '--', label='4 členy')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.gca().set_ylim([-.5,1.5])
plt.grid()
plt.legend()
plt.tight_layout()  # zabezpecuje spravny export popiskov osi grafu 
# plt.savefig('taylor.pdf')
```

### <font color=green> Veta 4 (Druhá resp. Lagrangeova veta o strednej hodnote) </font> 

Nech funkcia f je spojitá na intervale $\langle a,b\rangle$ a diferencovateľná na intervale $(a,b)$. Potom existuje $c \in (a,b)$ také, že $f'(c) = \dfrac{f(b) - f(a)}{b - a}$. 

---

---

#### Poznámka 3 

Veta 4 sa v angličtine niekedy označuje ako *Mean value theorem* a zvykne sa uvádzať i v nasledovnom znení: 

Nech $f\in C^1\langle a,b \rangle$, $x\in\langle a,b \rangle$, $y\in\langle a,b \rangle$ (BÚNV nech $x\le y$). Potom existuje $\xi\in\langle x,y\rangle\subseteq\langle a,b \rangle$ také, že platí 

$$f(y)-f(x)=f'(\xi)(y-x).$$

---

---

#### Poznámka 4 

* Geometrická interpretácia Lagrangeovej vety hovorí, že dotyčnica $t$ ku grafu funkcie $f$ v bode $[c,f(c)]$ je rovnobežná s úsečkou (spojnicou) spájajúcou body $[a,f(a)]$ a $[b,f(b)]$:
 
$$\text{Zdroj: Madaras (2019).}$$
Spomeňme si, že smernica priamky prechádzajúcej bodmi  $[a,f(a)]$, $[b,f(b)]$ je daná vzťahom $\dfrac{f(b) - f(a)}{b - a}$. Na druhej strane tiež vieme, že smernica dotyčnice ku grafu funkcie v bode $[c,f(c)]$ je rovná práve derivácii $f'(c)$. Podľa Lagrangeovej vety sa tieto smernice rovnajú, preto odpovedajúce priamky s týmito smernicami musia byť rovnobežné (sú nanajvýš posunuté jedna oproti druhej tak, ako to ilustruje obrázok).   


* Dôsledkom Vety 4 je tzv. **_Rolleho veta_**. Ide o špeciálny prípad, keď $f(a)=f(b)$. Presné znenie vrátane geometrickej interpretácie je možné nájsť napr. [tu](https://nbviewer.org/github/gajdosandrej/MTIb/blob/main/MTIb/Tema3/MTIbTema3DerivaciaFunkcie.ipynb?flush_cache=true) (úplný záver učebného textu).

$$\displaystyle\int_{0}^{x}\frac{1}{2}(x-t)^2e^t\,\mathrm{d}t\ge0,$$

$$e^x\ge1+x+\dfrac{x^2}{2}.$$

### <font color=green> Veta 5 (Taylorova pre funkciu dvoch premenných) </font> 

Nech funkcia $f(x,y)$ má spojité všetky parciálne derivácie až do rádu $n+1$ (vrátane) na oblasti $D=\{[x,y]:\,a\le x\le b,\,c\le y\le d\}$. Ďalej nech $[x_0,y_0]\in D$. Potom pre každý bod $[x,y]\in D$ existujú $\xi$ a $\eta$ také, že $x<\xi<x_0$, $y<\eta<y_0$ a súčasne platí 

$$f(x,y)=T_n(x,y)+R_n(x,y),$$ 

kde 

$$T_n(x,y)=f(x_0,y_0)+\left[(x-x_0)\dfrac{\partial f}{\partial x}(x_0,y_0)+(y-y_0)\dfrac{\partial f}{\partial y}(x_0,y_0)\right]+\ldots+\dfrac{1}{n!}\left[\displaystyle\sum_{j=0}^{n}{n\choose j}(x-x_0)^{n-j}(y-y_0)^j\dfrac{\partial^{n}f(x_0,y_0)}{\partial x^{n-j}\partial y^j}\right]$$ 

a 

$$R_n(x,y)=\dfrac{1}{(n+1)!}\left[\displaystyle\sum_{j=0}^{n+1}{n+1\choose j}(x-x_0)^{n+1-j}(y-y_0)^j\dfrac{\partial^{n+1}f(\xi,\eta)}{\partial x^{n+1-j}\partial y^j}\right].$$

#### <font color=blue> Príklad 4 </font> 

Dvojrozmerná Gaussova funkcia má predpis  

$$
    f(x,y) = \exp\left( -\frac{x^2+y^2}{b} \right). 
$$

Parciálne derivácie: 

\begin{align*}
    \dfrac{\partial f}{\partial x} &= -\frac{2x}{b} f(x,y) \\
    \dfrac{\partial f}{\partial y} &= -\frac{2y}{b} f(x,y) \\
    \\
    \dfrac{\partial^2f}{\partial x^2} &= \frac{2}{b} \left( \frac{2x^2}{b} - 1 \right) f(x,y) \\
    \dfrac{\partial^2f}{\partial y \partial x} = \dfrac{\partial^2f}{\partial x \partial y} &= \frac{4xy}{b^2} f(x,y) \\    
    \dfrac{\partial^2f}{\partial y^2} &= \frac{2}{b} \left( \frac{2y^2}{b} - 1 \right) f(x,y) \\
    \\
    \dfrac{\partial^3f}{\partial x^3} &= \frac{4x}{b^2} (3-2x^2 b) f(x,y) \\
    \dfrac{\partial^3f}{\partial x^2 \partial y} &= \frac{4y}{b^2} ( 1-2x^2 b ) f(x,y) \\
    \dfrac{\partial^3f}{\partial x \partial y^2} &= \frac{4x}{b^2} ( 1-2y^2 b ) f(x,y) \\
    \dfrac{\partial^3f}{\partial y^3} &= \frac{4y}{b^2} (3-2y^2 b) f(x,y) \\
\end{align*}

Aproximácie pomocou Taylorovho polynómu s rôznym počtom členov: 


```python
%matplotlib notebook  

import matplotlib.colors as colors
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

import numpy as np

def f(x,y,b):
    return np.exp(-(x**2+y**2)/b)

def dfdx(x,y,b):
    return -2*x/b * f(x,y,b)
def dfdy(x,y,b):
    return -2*y/b * f(x,y,b)

def d2fdx2(x,y,b):
    return 2/b * (2*x**2/b - 1) * f(x,y,b)
def d2fdxdy(x,y,b):
    return 4*x*y/b**2 * f(x,y,b)
def d2fdy2(x,y,b):
    return 2/b * (2*y**2/b - 1) * f(x,y,b)

def d3fdx3(x,y,b):
    return 4*x/b**2 * (3 - 2*x**2*b) * f(x,y,b)
def d3fdx2dy(x,y,b):
    return 4*y/b**2 * (1 - 2*x**2*b) * f(x,y,b)
def d3fdxdy2(x,y,b):
    return 4*x/b**2 * (1 - 2*y**2*b) * f(x,y,b)
def d3fdy3(x,y,b):
    return 4*y/b**2 * (3 - 2*y**2*b) * f(x,y,b)

x = np.arange(-1,1,0.1)
y = np.arange(-1,1,0.1)
x, y = np.meshgrid(x,y)

b  = 1.0
x0 = 0.1
y0 = 0.2

t1 = f(x0,y0,b)
t2 = dfdx(x0,y0,b)*(x-x0) + dfdy(x0,y0,b)*(y-y0)
t3 = 1/2 * ( d2fdx2(x0,y0,b)*(x-x0)**2 + d2fdxdy(x0,y0,b)*(x-x0)*(y-y0) + d2fdy2(x0,y0,b)*(y-y0)**2 )
t4 = 1/6 * ( d3fdx3(x0,y0,b)*(x-x0)**3 + d3fdx2dy(x0,y0,b)*(x-x0)**2*(y-y0) + d3fdxdy2(x0,y0,b)*(x-x0)*(y-y0)**2 + d3fdy3(x0,y0,b)*(y-y0)**3 )

z = f(x,y,b)  # the actual result

z2 = t1+t2    # 2 terms
z3 = z2+t3    # 3 terms
z4 = z3+t4    # 4 terms

# plot up the results
# fig = plt.figure(figsize=plt.figaspect(0.25))
fig = plt.figure(figsize=(8,2))
norm = colors.Normalize(vmin=z.min(), vmax=z.max())
cmap = plt.cm.gist_heat

ax = fig.add_subplot(1, 4, 1, projection='3d')
# ax.set_xlabel('x')
# ax.set_ylabel('y')
ax.set_title('Exact')
surf = ax.plot_surface( x, y, z, cmap=cmap, norm=norm )

ax = fig.add_subplot(1, 4, 2, projection='3d')
# ax.set_xlabel('x')
# ax.set_ylabel('y')
ax.set_title('2-term TS')
surf = ax.plot_surface( x, y, z2, cmap=cmap, norm=norm )

ax = fig.add_subplot(1, 4, 3, projection='3d')
# ax.set_xlabel('x')
# ax.set_ylabel('y')
ax.set_title('3-term TS')
surf = ax.plot_surface( x, y, z3, cmap=cmap, norm=norm )

ax = fig.add_subplot(1, 4, 4, projection='3d')
# ax.set_xlabel('x')
# ax.set_ylabel('y')
ax.set_title('4-term TS')
surf = ax.plot_surface( x, y, z4, cmap=cmap, norm=norm )
fig.colorbar(surf)

# plt.savefig('taylor2d.pdf')
plt.show()
```

 # <font color=brown> Rýchlosť konvergencie, asymptotika</font>

## Big $\mathcal{O}$, little $\mathcal{o}$

V numerickej matematike častokrát študujeme **rýchlosť alebo mieru s akou rastie či klesá chyba výpočtu** (aproximácie). Napr. keď aproximujeme $e^h$ Taylorovým polynómom prvého stupňa na okolí bodu $x=0$, tak dostaneme 

$$e^h-(1+h)=\frac{1}{2}h^2e^{\xi},$$ 

pričom $\xi$ je neznáma veličina medzi $0$ a $h$. Aj keď nepoznáme presne $e^\xi$, vieme, že pre $h$ blízke $0$, je to takmer konštantný výraz (v tomto prípade približne $1$) a teda chyba (aproximácie) $e^h-(1+h)$ je zhruba úmerná (proporcionálna) $h^2$ (pre malé $h$). Inými slovami, chyba aproximácie $e^h-(1+h)$ sa správa približne ako funkcia $h^2$ pre malé hodnoty $h$. Poznať túto približnú úmernosť (proporcionalitu) je zvyčajne dôležitejšie ako poznať samotnú "pomaly sa meniacu konštantu" $e^\xi$. Označenia $\mathcal{O}$ (big "$\mathcal{O}$") a $\mathcal{o}$ (small "$\mathcal{o}$") sa používajú na **popis uvedenej približnej proporcionality resp. asymptotického správania** (napr. chyby aproximácie). Toto označenie zaviedol nemecký matematik Edmund Landau, pracujúci v oblasti teoretickej aritmetiky. 

Symbolika, ktorú ďalej exaktne popíšeme slúži v podstate na to, aby sme mohli charakterizovať rýchlosť zmeny funkcie (t.j. miera rastu/klesania) pri meniacom sa argumente funkcie (zväčsovanie/zmenšovanie argumentu funkcie) a to tak, že nebudeme musieť "ísť príliš do detailov". 

---

### <font color=red> Definícia 1 (Big $\mathcal{O}$, little $\mathcal{o}$) </font> 

Nech $\{x_k\}$, $\{\alpha_k\}$ sú dve (číselné) postupnosti. Ak existujú konštanty $c$, $r$ také, že $|x_k|\le c|\alpha_k|$ pre $k\ge r$, tak píšeme $x_k=\mathcal{O}(\alpha_k)$ pre $k\to\infty$. Ďalej, ak k ľubovoľnému danému $\varepsilon>0$ existuje také $r$, že $|x_k|<\varepsilon|\alpha_k|$ pre $k\ge r$, tak zapíšeme $x_k=\mathcal{o}(\alpha_k)$. Podobne, ak $f(h)$, $g(h)$ sú spojité funkcie premennej $h$, označíme $f(h)=\mathcal{O}(g(h))$ pre $h\to0$, keď existujú konštanty $c$, $\delta$ také, že $|f(h)|\le c|g(h)|$ pre $|h|<\delta$. 

--- 

---

#### Poznámka 5 

* **Všimnime si**, že ak $x_k=\mathcal{O}(\alpha_k)$, potom $\lim\limits_{k\to\infty}\left|\dfrac{x_k}{\alpha_k}\right|\le c.$ Podobne, ak $x_k=\mathcal{o}(\alpha_k)$, potom $\lim\limits_{k\to\infty}\left|\dfrac{x_k}{\alpha_k}\right|= 0.$


* Znak **$\mathcal{O}$** v podstate **označuje rád**. V ilustračnom príklade pred Definíciou 1 by to znamenalo, že ak chybu aproximácie označíme $f(h)=e^h-(1+h)$, potom platí $f(h)=\mathcal{O}(h^2)$ a hovoríme, že "$f$ vykazuje konvergenciu k $0$ rádu $2$ pre $h$ idúce k $0$". 


* Označenie $\mathcal{O}$ sa často používa v kontexte analýzy chýb pri vektorovo-hodnotových funkciách (napr. funkcie z $\mathbb{R}$ do $\mathbb{R}^{m}$). Nech $e(h)$ je vektor reprezentujúci chybu iného vektora; potom zapisujeme 

$$e(h)=\mathcal{O}(h^k) \text{ pre } h\to\infty,$$ 

ak existuje konštanta $c$ závisiaca od $h$ taká, že $\lVert e(h)\rVert\le ch^k$ pre všetky $h\in\langle0,h_0\rangle$. Zhruba povedané, pre dostatočne malé $h$, norma $\lVert e(h)\rVert$ nejde k $0$ pomalšie ako $h^k$.

Preto 

$$e^{1/k}-1-1/k\le\frac{1}{k^2}\left(\frac{e^1}{2}\right) \text{ pre } k\ge1.$$

### <font color=red> Definícia 2 </font> 

Nech $\lim\limits_{h\to0}f(h)=L$. Zapisujeme $f(h)-L=\mathcal{O}(g(h))$, ak existuje konštanta $K$ taká, že $\dfrac{\left|f(h)-L\right|}{\left|g(h)\right|}<K$ pre dostatočne malé $h>0$.

#### <font color=blue> Príklad 6 </font> 

Ukážte, že $\cos h - 1 + \frac{h^2}{2}=\mathcal{O}(h^4)$. 

<ins> Riešenie: </ins>

Podľa Taylorovej vety máme 

$$\cos h = 1 - \frac{h^2}{2}+\frac{1}{3!}\displaystyle\int_{0}^{h}\cos(t)(h-t)^3\,\mathrm{d}{t}.$$ 

Preto 

$$\left|\cos h-1+\frac{h^2}{2}\le\frac{1}{6}\displaystyle\int_{0}^{h}(h-t)^3\,\mathrm{d}{t}=\frac{h^4}{24}\right|.$$

Odkiaľ 

$$\dfrac{\left|\cos h-1+\frac{h^2}{2}\right|}{\left|h^4\right|}\le\dfrac{1}{24}.$$

#### Poznámka 6 

* Zápis $f(h)=\mathcal{O}\left(g(h)\right)$ v podstate vyjadruje nasledovného: existuje $c>0$ také, že $\left|f(h)\right|\le c\left|g(h)\right|$ pre $h\to a$, čo je to isté, ako $\lim\limits_{h\to a}\left|\dfrac{f(h)}{g(h)}\right|\le c.$. Uvedená symbolika hovorí o tom, že od určitého bodu funkcie $f$, $g$ rastú zhruba rovnako rýchlo - až na násobok konštantou (rádovo rovnaká rýchlosť rastu). 

 

* Zvyčajne sa uvažuje $a=0$ alebo $a=\pm\infty$. Ak je $a=\infty$, potom existujú $h_0$ a $c$ také, že pre $h>h_0$ platí vzťah resp. nerovnosť v predošlom bode. Pokiaľ je $a=0$, tak existujú $h_0$ a $c$ také, že nerovnosť uvedená v predchádzajúcom bode platí pre $|h|<|h_0|$. 


* Zavedené označenie $f(h)=\mathcal{O}\left(g(h)\right)$ vlastne vyjadruje, že funkcia $f$ rastie nanajvýš tak rýchlo ako funkcia $g$ (až na konštantný faktor/násobok), keď $h\to a$ (alternatívne, ak $g\to0$, tak hovoríme, že $f$ klesá aspoň tak rýchlo ako $g$). 


* Označenie $f(h)=\mathcal{o}(g(h))$ znamená, že ku každému $\varepsilon>0$ existuje $h_0$ také, že pre $h>h_0$ platí $|f(h)|<\varepsilon|g(h)|$, čo je to isté ako $\lim\limits_{h\to\infty}\left|\dfrac{f(h)}{g(h)}\right|= 0$. Uvedená symbolika vyjadruje, že funkcia $g$ rastie oveľa (resp. rádovo) rýchlejšie než funkcia $f$. 


* Rovnaké definície a interpretácie platia i v prípade (číselných) postupností, kde však jediná zmysluplná limita je $n\to\infty$.

## Rýchlosť konvergencie 

Jeden zo spôsobov, ako porovnávať kvalitu resp. **efektívnosť algoritmov**, je rýchlosť ich konvergencie k nejakej limitnej hodnote. Pre ilustráciu uvažujme interačný algoritmus, ktorý sa snaží nájsť maximum/minimum určitej funkcie a chceme odhadnúť, ako dlho bude trvať dosiahnutie tejto optimálnej hodnoty. Za týmto účelom predstavíme štyri typy rýchlostí konvergencie – *sublineárna*, *lineárna*, *superlineárna* a *kvadratická* – sú zoradené od najpomalšej po najrýchlejšiu. 

V našom kontexte bude rýchlosť konvergencie algoritmu zvyčajne určená tým, koľko informácií o účelovej funkcii $f$ sa používa v procese aktualizácie algoritmu. Také algoritmy, ktoré využívajú málo informácií o funkcii $f$, ako je napr. [algoritmus bisekcie](https://pythonnumericalmethods.berkeley.edu/notebooks/chapter19.03-Bisection-Method.html), konvergujú pomaly. Naproti tomu algoritmy, ktoré vyžadujú viac informácií o $f$, napr. v podobe derivácií uvažovanej funkcie, konvergujú rýchlejšie - napr. [Newtonov-Raphsonov algoritmus](https://pythonnumericalmethods.berkeley.edu/notebooks/chapter19.04-Newton-Raphson-Method.html). 

Pre jednoduchosť i názornosť si ďalej predstavme, že uvažovaný iteračný algoritmus (napr. pre hľadanie minima/maxima funkcie) je možné vyjadriť ako číselnú postupnosť $\left\{x_n\right\}$. Presnejšie teda výsledok výpočtu každého kroku resp. iterácie takého algoritmu, je možné reprezentovať hodnotou $x_n$ uvedenej postupnosti. Konkrétne teda hodnota $x_5$ môže napr. vyjadrovať aproximáciu minima/maxima $x^*$ zadanej funkcie, pričom táto aproximácia je výsledkom výpočtu piatej iterácie. Tu vznikajú prirodzené a dôležité otázky vyplývajúce z praxe: **Ako charakterizovať rýchlosť, ktorou postupnosť $\left\{x_n\right\}$ (resp. príslušný algoritmus) konverguje k $x^*$ ?** Predpokladajme, že vieme vygenerovať postupnosť $x_1,x_2,\ldots$ pomocou iteračného algoritmu a chceme získať výsledok $x_n$ (t.j. aproximáciu hodnoty $x^*$) s presnosťou na $k$ číslic ($x_n\approx x^*$); **koľko iterácií (algoritmu) je potrebných vykonať na dosiahnutie požadovanej presnosti?** (tzn. zredukovať chybu aproximácie faktorom $10^{-k}$); Pokiaľ máme k dispozícii výpis chýb zodpovedajúcich jednotlivým iteráciám, **ako je možné určiť/odhadnúť rýchlosť konvergencie postupnosti (resp. algoritmu) $\left\{x_n\right\}$ približných riešení?** 

Chybu $k$-tej iterácie (t.j. vzdialenosť aproximácie $x_k$ od skutočnej hodnoty $x^*$) je možné vyjadriť vzťahom $\varepsilon_k=\left|x_k-x^*\right|$. Zrejme postupnosť $\left\{x_n\right\}$ (resp. príslušný algporitmus) konverguje práve vtedy, keď $\lim\limits_{k\to\infty}\varepsilon_k=0$. V praxi často nepoznáme presnú (resp. hľadanú/optimálnu) hodnotu $x^*$, preto môžeme napr. "useknúť" algoritmus resp. odpovedajúcu postupnosť po $N$ krokoch a urobiť odhad $\tilde{\varepsilon}_k=|x_k-x_N|$. Alternatívou môže byť sledovanie $\delta_k=|x_k-x_{k-1}|$. Pokiaľ algoritmus konverguje, potom očakávame tiež $\delta_k\to0$ no a v prípade, že je $\delta_k$ dostatočne malé, tak je možné výpočet (algoritmus) ukončiť.

### <font color=red> Definícia 3 </font> 

Nech (číselná) postupnosť $\{x_k\}$ má limitu $x^*$. Ak existujú konštanty $C$, $p$ a prirodzené číslo $N$ také, že pre každé $k\ge N$ platí $\left|x_{k+1}-x^*\right|\le C\left|x_{k}-x^*\right|^p$, tak hovoríme, že **_rýchlosť konvergencie_** je **rádu aspoň** $p$. Ak $p=1$ a $C<1$, rýchlosť konvergencie sa označuje ako **_lineárna_**. Pokiaľ $p=1$ a $C=1$, hovoríme o **_sublineárnej_** konvergencii. Ak $p=2$, potom ide o **_kvadratickú_** rýchlosť konvergencie.  

--- 

---

#### Poznámka 9

* Rýchlosť konvergencie rádu $p$ sa niekedy definuje priamo ako limita v zmysle $\lim\limits_{k\to\infty}\dfrac{\left|x_{k+1}-x^*\right|}{\left|x_k-x^*\right|^p}=C$, pričom ak sa v limite nahradí rovnosť nerovnosťou $\le$, tak ide o konvergenciu rádu aspoň $p$. **POZOR!** Konštanta $C$ uvedená tu a v Definícii 3 nemusí byť úple tá istá (avšak v oboch prípadoch pri lineárnej konvergencii musí byť z intervalu $(0,1)$). 


* Zjednodušene povedané, v prípade lineárnej konvergencie ($0<C<1$), vzťah $\left|x_{k+1}-x^*\right|\approx C\left|x_k-x^*\right|$ pre $n\to\infty$ vyjadruje, že v každom kroku sa chyba (aproximácie) znižuje o faktor $C$, keď sa postupnosť (aproximácií) blíži k limite. 


* Definícia 3 pojednáva o tzv. *Q-konvergenciách*, no nepokrýva špeciálne postupnosti, napr. také, ktoré síce konvergujú k nule avšak zahŕňajú v sebe určité "oscilácie" reps. rýchlosť ich konvergencie je premenlivá ... z tohto dôvodu sa niekedy zavádza tzv. *[R-konvergencia](https://en.wikipedia.org/wiki/Rate_of_convergence)*.  


* Prirodzene, keď uvažujeme miesto postupnosti skalárov postupnosť vektorov, tak absolútna hodnota v Definícii 3 sa nahradí príslušnou normou vektora.

#### Poznámka 10 

Kvadratická konvergencia je rýchla. Pre Príklad 4 máme: ak $|x_k-2|=0.01$, tak potom $|x_{k+1}-2|\le\frac{1}{4}(0.01)^2$, $|x_{k+2}-2|\le\frac{1}{64}(0.01)^4$, ... 

```python
# rekurzivne definovana postupnost {x_k} z Prikladu 4
def xk(k): 
    if k == 0: 
        return 3 
    a = xk(k-1)**2 + 4 
    b = xk(k-1)*2 
    return a/b  
```

### <font color=red> Definícia 4 </font> 

Je daná postupnosť $\{x_k\}$ s limitou $x^*$. Hovoríme, že táto postupnosť má **_superlineárny rád konvergencie_**, akk platí 

$$\dfrac{\left|x_{k+1}-x^*\right|}{\left|x_k-x^*\right|}\to0 \text{ pre } k\to \infty.$$

---

#### Poznámka 11 

Superlineárna konvergencia je rýchlejšia ako lineárna ale v princípe môže byť pomalšia ako konvergencia rádu $\alpha$, kde $\alpha>1$. 

---

#### <font color=blue> Príklad 10 </font> 

Postupnosť $\{x_k\}$ definovaná vzťahom $x_k=\dfrac{1}{2^{k^2}}$ superlineárne konverguje k $0$, keďže 

$$\dfrac{\left|x_{k+1}-0\right|}{\left|x_k-0\right|}=\dfrac{2^{k^2}}{2^{(k+1)^2}}=\dfrac{1}{2}\cdot\dfrac{1}{2^{2k}}\to0 \text{ pre } k\to\infty.$$

Na druhej strane ale daná postupnosť nekonverguje kvadraticky lebo 

$$\dfrac{\left|x_{k+1}-0\right|}{\left|x_k-0\right|^2}=\dfrac{1/2^{(k+1)^2}}{1/\left(2^{k^2}\right)^2}=2^{k^2-2k-1}\not\to0 \text{ pre } k\to\infty.$$

Dokonca je možné ukázať, že postupnosť $\{x_k\}$ nedosahuje stupeň konvergencie $\alpha$ pre žiadne $\alpha>1$.

#### <font color=blue> Príklad 11 </font>

```python
# generovanie kvadraticky konvergentnej postupnosti 
n = 9
C = 2.0
p = 2
eps2 = [0.2]
for i in range(n):
    eps2.append(C * eps2[-1]**p)

plt.figure(figsize=(4,3))
plt.semilogy(eps2)
plt.xlabel('k')
plt.ylabel(r"$\epsilon$")
plt.title("Kvadratická konvergencia")
plt.show()
```

```python
# generovanie sub-linearne (logaritmicky) konvergentnej postupnosti
n = 50
eps3 = []
for k in range(1,n):
    eps3.append(1.0/k)

plt.figure(figsize=(4,3))
plt.semilogy(eps3)
plt.xlabel('k')
plt.ylabel(r"$\epsilon$")
plt.title("Sublineárna konvergencia")
plt.show()
```

```python
# vsetky tri typy konvergencie v jednom obrazku 
plt.figure(figsize=(6,4))
plt.semilogy(eps1, label='lineárna')
plt.semilogy(eps2, label='kvadratická')
plt.semilogy(eps3, label='sublineárna')
plt.xlabel('k')
plt.ylabel(r"$\epsilon$")
plt.ylim(1e-16, 1e0)
plt.title("Typy konvergencie")
plt.legend()
plt.show()
```

--- 

Existujú rôzne prístupy **ako odhadnúť rád konvergencie $p$**. Stručne popíšeme myšlienku jedného z nich, ktorý celkom slušne funguje - svedčí o tom kód uvedený nižšie. Princíp je nasledovný. Predpokladajme, že je k dispozícii (konečná) postupnosť chýb $\left\{\varepsilon_k\right\}$ (podobne ako v Príklade 11). Z Definície 3 máme vzťah $\varepsilon_{k+1}=C\varepsilon_k^p$, kde $p$ je rád rýchlosti konvergencie, ktorý chceme odhadnúť. Logaritmovaním uvedeného rekurentného vzťahu (diskrétny dynamický systém resp. diferenčná rovnica) a zavedením označenia $z_k=\ln(\varepsilon_k)$ dostaneme $z_{k+1}=pz_{k}+\ln(C)$. Riešením tejto diferenčnej rovnice štandardnou technikou s dodatočnou počiatočnou podmienkou $z_0=0$ získame $z_k=\dfrac{p^k\ln(C)}{p-1}-\dfrac{\ln(C)}{p-1}$. Odčítaním $z_k$ od $z_{k+1}$, logaritmovaním, následnou úpravou a spätnou substitúciou dôjdeme k vzťahu $\ln\left|\ln\left(\dfrac{\varepsilon_{k+1}}{\varepsilon_k}\right)\right|=k\ln{p}$. Ak položíme $y_k=\ln\left|\ln\left(\dfrac{\varepsilon_{k+1}}{\varepsilon_k}\right)\right|$, $x_k=k$, potom $\ln(p)$ sa dá odhadnúť metódou najmenších štvorcov (spomeňte si na odvodenie regresnej priamky z UAD). Na záver si už len stačí uvedomiť, že $e^{\ln(p)}=p$. 


```python
def estimate_p(eps):
    """
    estimate rate of convergence p from sequence eps 
    """
    x = np.arange(len(eps) -1)
    y = np.log(np.abs(np.diff(np.log(eps))))
    line = np.polyfit(x, y, 1) # fit degree 1 polynomial
    p = np.exp(line[0]) # find p
    return p
```

```python
print(estimate_p(eps1)) # linear
print(estimate_p(eps2)) # quadratic
print(estimate_p(eps3)) # sublinear
```

\end{aligned}$$

$$\begin{aligned}
    f'(x) &= -\frac{1}{x^2} \\

$$\begin{aligned}

c) Aproximujte funkciu $f(x) = e^{-x}$ v okolí bodu $x_0 = 0.5$ postupne Taylorovým polynómom prvého až piateho rádu. Znázornite chybu aproximácie zadanej funkcie pomocou jednotlivých polynómov a to vzhľadom na vzdialenosť od bodu $x_0=0.5$, v ktorom sú jednotlivé polynómy skonštruované. 

<ins> Riešenie: </ins>

Nasledujúci graf zobrazuje chybu aproximácie ako funkciu diferencie $\left|\Delta x\right|=\left|x-x_0\right|$, pričom tieto diferencie sú logaritmicky rozložené v intervale $\langle-2,1\rangle$. Niekedy sa kvôli lepšej prehľadnosti na zobrazenie používa i tzv. [log-log graf](https://en.wikipedia.org/wiki/Log%E2%80%93log_plot) - najmä pri veľkých (rádových) rozdieloch v chybách a/alebo v "krokoch". 

```python
%matplotlib inline
from matplotlib import rcParams
from math import factorial

rcParams.update({'font.size': 14})

x0 = 0.5
dx = np.logspace(-2,1,100)
x = x0+dx
exact = np.exp(-x)

n = np.arange(0,5)
terms = np.array([(x-x0)**i/factorial(i) * (-1)**i * np.exp(-x0) for i in n ])

plt.figure(figsize=(6,4))
for i in n:
    taylor = np.sum( terms[0:i+1,:], axis=0 )
    plt.plot(dx,np.abs(exact-taylor),label=str(i+1)+' terms')   
    # plt.loglog(dx,np.abs(exact-taylor),label=str(i+1)+' terms') 

# plot reference slopes
xx = np.logspace(-1,0,2)
plt.loglog(xx,xx,'ko-',label='slope=1')
plt.loglog(xx,0.1*xx**3,'ko-',label='slope=33')
plt.loglog(xx,0.001*xx**5,'ko-',label='slope=5')

plt.xlabel('$|\Delta x|$')
plt.ylabel('$\epsilon$ (truth - approximation)')
plt.legend(fontsize='x-small')
plt.grid(True)
plt.title('Taylor series of $e^{-x}$')
plt.tight_layout()  # ensures that the axis labels export properly
# plt.savefig('taylor-exp-converge.pdf')
```

Môžeme si všimnúť, že so zmenou rádu/stupňa $n$ Taylorovho polynómu sa mení tiež sklon chybovej funkcie na logaritmickej škále (_log-log scale_). Na druhej strane i zmena $\Delta x$ ovplyvňuje presnosť. V konečnom dôsledku teda miera, s akou sa mení chyba aproximácie je priamo závislá na oboch, t.j. na $\Delta x$ aj $n$. Hovoríme tomu rád presnosti metódy a v tomto prípade napr. konkrétne pre $n=2$ máme 

$$
T(x) = f(x_0) + \Delta x f^\prime(x_0) + \frac{\Delta x^2}{2} f^{\prime\prime}(x_0) + \mathcal{O}\left(\Delta x^3\right). 
$$
