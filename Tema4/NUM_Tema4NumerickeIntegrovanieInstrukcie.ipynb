# <font color=brown> Motivácia </font>

**Počítačom podporovaná výroba** závisí od presného riadenia pohybu po predpísanej dráhe. Napríklad sústruhy alebo frézky pod počítačovým numerickým riadením (ang. *Computer Numerical Control* - CNC) sa spoliehajú na **parametrické krivky**, ktoré sú často dané **kubickými** alebo **Bézierovými splajnmi** z počítačom podporovaného konštrukčného softvéru, na opis dráhy rezných alebo tvarovacích nástrojov. **Počítačom generované animácie** vo filmovej tvorbe, v počítačových hrách i v aplikáciách virtuálnej reality čelia podobným problémom. 

 

**Počítačom podporované modelovanie a výroba** si vyžaduje presné riadenie priestorovej polohy pozdĺž predpísanej dráhy pohybu. Tu sa uplatňuje **numerická kvadratúra** na vyriešenie základnej časti problému: získanie ekvipartície resp. rozdelenie ľubovoľnej cesty (dráhy) na rovnako dlhé podcesty. <br/> 
Pri **problémoch obrábania kovov**, dreva či iných materiálov pomocou tzv. **počítačového numerického riadenia** (ang. *Computer Numerical Control* - CNC) je výhodné udržiavať konštantnú rýchlosť pozdĺž dráhy. Počas každej sekundy by sa mal dosiahnuť pokrok pozdĺž rovnakej dĺžky rozhrania materiálu stroja. V iných aplikáciách plánovania pohybu, vrátane počítačovej animácie, môžu byť potrebné komplikovanejšie krivky postupu: Ručné natiahnutie sa po kľučke dverí môže začínať a končiť nízkou rýchlosťou, no medzi tým môže byť rýchlosť vyššia. Robotika i aplikácie virtuálnej reality vyžadujú konštrukciu parametrizovaných kriviek a plôch, po ktorých sú navigované. Vytvorenie tabuľky malých rovnakých prírastkov pozdĺž dráhy je často nevyhnutným prvým krokom.

 

Integrály ako také sa vyskytujú v rôznych oblastiach ľudskej činnosti, od **financií** cez **inžinierstvo** a rôzne **prírodné vedy** až po **humanitné vedy**. Nájdeme ich tiež v rôznych **fyzikálnych modeloch** reálneho sveta ale aj v **modelovaní virtuálneho sveta** - napr. v podobe tzv. (fyzikálnych) egine-och viedohier. Vo všeobecnosti platí, že čím lepší [engine](https://en.wikipedia.org/wiki/Physics_engine), tým realistickejšia videohra. Integrály vyskytujúce sa v praxi sú veľakrát zložité a tým pádom analyticky veľmi ťažko riešiteľné, ak vôbec. Tu prichádza na pomoc **numerické integrovanie**, ktoré umožňuje získať dosť presné výsledky zložitých integrálov a pomáha pri riešení praktických problémov. <br/> 
V **teórii pravdepodobnosti** i v **matematickej štatistike** sa pri modelovaní a popise reálnych fenoménov tiež vyskytujú zložité integrály, ktoré sú bez pomoci chytrých **numerických algoritmov** prakticky neriešiteľné. 

V tejto téme sa budeme zaoberať **rôznymi pravidlami pre numerické integrovanie**. Konkrétne pôjde o **jednoduché kvadratúrne formule - lichobežníková, obdĺžniková, Simpsonova**. Názorne si vysvetlíme i tzv. **zložené kvadratúrne formule**, ktoré môžu vzniknúť z uvedených jednoduchých. Pozrieme sa na tieto pravidlá i všeobecnejšie cez tzv. **Newtonove-Cotesove vzorce**, ktoré vznikajú integrovaním Lagrangeovho interpolačného polynómu danej funkcie resp. integrandu. Zaujímavou a efektívnou metódou je **Rombergova integrácia**, založená na kombinácii lichobežníkovej formuly a Richardsonovej extrapolácie. Inou alternatívou bude **Gaussova kvadratúra** a rôzne jej špeciálne typy ako napr. **Gaussova-Legendrova** či **Gaussova-Hermitova kvadratúra** a ďalšie. Pri tomto type kvadratúr budú hrať dôležitú úlohu rôzne triedy **ortogonálnych polynómov**. Okrem samotných vzorcov pre numerické integrovanie preskúmame aj **chyby** vznikajúce **pri aproximácii integrálov**. Samozrejme, nebudú chýbať ilustračné príklady výpočtov v Pythone.

* Porozumieť princípom vybraných numerických kvadratúr a poznať ich vlastnosti. 


* Definovať a overiť (algebraický) rád presnosti kvadratúrnej formule. 


* Definovať a overiť ortogonalitu funkcií na danom intervale. 


* Vedieť aplikovať vhodné moduly Python-u na približný výpočet integrálov pomocou vybraných numerických kvadratúr. 


* Rozumieť princípu odvádzania (diskretizačných) chýb vybraných numerických kvadratúr. 


* Odhadnúť chybu aproximácie (pomocou vzorca) v konkrétnom príklade numerického integrovania. 


* Vysvetliť podmienenosť úlohy numerického integrovania. 

### Pokročilé 


* Vedieť naprogramovať metódy (z učebného textu) pre numerické integrovanie. 


* Aplikovať vhodnú numerickú kvadratúru na zadaný integrál a pokiaľ je to možné, porovnať numerické riešenie s analytickým (napr. určiť relatívnu chybu aproximácie). 


* Nájsť koeficienty (váhy, prípadne uzly) zadanej kvadratúrnej formule tak, aby spĺňala určité požiadavky. 



* Určiť počet uzlov (podintervalov) danej kvadratúry tak, aby chyba aproximácie konkrétneho integrálu neprekročila stanovenú hranicu.


* Transformovať integrál vhodnou substitúciou a následne aplikovať vhodnú numerickú kvadratúru na jeho aproximáciu. 



* Aplikovať vhodnú numerickú kvadratúru na vyriešenie úlohy s kontextom.
