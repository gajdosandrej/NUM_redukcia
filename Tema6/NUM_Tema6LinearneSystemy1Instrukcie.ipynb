# <font color=brown> Motivácia </font>

Fyzikálne zákony riadia každú vybudovanú stavbu, od mrakodrapov a mostov až po skokanské mostíky a lekárske zariadenia. Statické a dynamické zaťaženie spôsobuje deformáciu alebo ohýbanie materiálov. **Matematické modely ohýbania** sú základnými nástrojmi statika. Miera, do akej sa konštrukcia pri zaťažení ohýba, závisí od tuhosti materiálu meranej pomocou tzv. Youngovho modulu. Vzťah medzi napätím a tuhosťou je modelovaný diferenciálnou rovnicou, ktorá sa po diskretizácii redukuje na riešenie sústavy lineárnych rovníc. <br/>
Na zvýšenie presnosti sa používa jemná diskretizácia, vďaka čomu je systém lineárnych rovníc veľký a zvyčajne riedky. **Gaussove eliminačné metódy** sú účinné pre stredne veľké matice, ale pre veľké, riedke systémy sú potrebné **špeciálne iteračné algoritmy**. 

 

V ekonómii existuje tzv. **vstupno-výstupná analýza**. Konkrétne ide o to, že ekonómovia chcú zistiť stav rovnováhy medzi dopytom a ponukou určitého tovaru. Je možné uvažovať tzv. výrobný model, v ktorom $m \ge n$ tovární (alebo výrobných liniek) vyrába $n$ rôznych produktov. Musia čeliť vnútornému dopytu po tovare (vstup) potrebnom pre továrne na ich vlastnú výrobu, ako aj vonkajšiemu dopytu (výstupu) spotrebiteľov. Leontief navrhol lineárny výrobný model, pre ktorý je množstvo určitého výstupu úmerné množstvu použitého vstupu. Naznačené modelovanie a následné hľadanie spomínanej rovnováhy vedie k **riešeniu sústavy lineárnych rovníc**. 

 

V téme 6 predstavíme tzv. **priame metódy na riešenie sústav lineárnych rovníc**. Konkrétne pôjde o **naivnú Gaussovu eliminačnú metódu**, riešenie sústavy s využitím **LU rozkladu** či **Choleskeho dekompozície** matice sústavy. Poukážeme na možné problémy vznikajúce počas riešenia a budeme sa ich snažiť odstrániť pomocou chytrejších modifikácií uvedených metód - napr. implementovaním **výberu hlavných prvkov (pivotov)**. Naznačíme, ako je možné celý proces riešenia zjednodušiť pri špeciálnych maticiach sústav (**trojuholníkové** či **pásové matice**). Ukážeme si, ako je možné využiť **Gaussovu elimináciu**  na **výpočet determinatu** či na **riešenie sústav s viacerými vektormi pravých strán**.  Nevynecháme ani **analýzu chýb** resp. **analýzu podmienenosti**. Všetko bude doplnené ilustračnými príkladmi, obrázkami a kódmi.
