# <font color=brown> Motivácia </font>

Objav **dvojitej špirálovej štruktúry** DNA v roku 1953 viedol o polstoročie neskôr k takmer úplnému **sekvenovaniu ľudského genómu**. Sekvencia v sebe nesie zakódované informácie o skladaní reťazcov aminokyselín do jednotlivých proteínov, ktoré vykonávajú prejavy života. Tieto informácie potrebujú preklad (dekódovanie), aby ich bolo možné nasmerovať k podrobnému porozumeniu fyziologickej funkcie. Rad potenciálnych aplikácií vrátane **génovej terapie i návrhu liečiv môže podporovať** včasnú prevenciu, diagnostiku a liečenie choroby. Skladanie aminokyselín do funkčných proteínov zásadne závisí od van der Waalsových síl, mikroskopickej príťažlivosti a odporu medzi neviazanými atómami. Pri štúdiu modelov atómových klastrov, v ktorých sú tieto sily modelované potenciálom Lennard-Jones, sú kľúčové konfigurácie s minimálnou energiou, preto na scénu prichádza **optimalizácia** a stáva sa potrebnou súčasťou naznačeného výskumu ... 

 


Ďalšia zaujímavá, praktická úloha, kde **optimalizácia v kombinácii so strojovým učením** uľahčuje život resp. zefektívňuje procesy je nasledovná: Predstavte si veľkú databázu (rádovo v stotisícoch) textov v elektronickej podobe (básne, poviedky, odborné texty atď.) a tieto texty je potrebné automatizovaným spôsobom spoľahlivo roztriediť do kategórií (napr. podľa žánrov) tak, aby sa to nemuselo realizovať "ručne" resp. aby to reálne nemusel prechádzať a triediť človek. 

 


**Rozpoznanie obrazu** (ang. _image recognition_) je schopnosť identifikovať napríklad objekty, miesta, ľudí, text alebo činnosť na obrázku. Rozpoznanie obrazu je široko používané v strojovo orientovaných vizuálnych úlohách, ako je napríklad označovanie obsahu obrázkov tagmi, vyhľadávanie obrázkov na základe obsahu, navigovanie robotov a riadenie autonómnych áut. Algoritmy na rozpoznávanie obrázkov sú väčšinou založené na **hlbokom strojovom učení** a najlepšie výsledky dosahujú **konvolučné neurónové siete**. Práve učiaci resp. trénovací proces neurónových sietí sa nezaobíde bez **efektívnej numerickej optimalizácie**. 

 

V prípade **minimalizácie funkcie jednej reálnej premennej** predstavíme **metódy intervalovej redukcie** a to konkrétne **modifikáciu metódy polenia intervalov, metódu zlatého rezu. Z trochu iného súdka budú **metóda kvadratickej (parabolickej) interpolácie** i **Newtonova metóda**. Pokiaľ ide o **hľadanie minima funkcie viacerých premenných**, tak z kategórie **komparatívnych metód** bude prezentovaná **Nelderova-Meadova metóda (resp. metóda simplexov)** a spomedzi **gradientných metód** sa pozrieme na **metódu najväčšieho spádu, Newtonovu metódu** a spomenieme tiež niektoré **kvázi-Newtonovske metódy**. Objasníme súvis niektorých uvedených metód s riešením sústav (ne)lineárnych rovníc.
