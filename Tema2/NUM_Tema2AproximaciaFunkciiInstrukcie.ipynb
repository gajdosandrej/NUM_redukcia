# <font color=brown> Motivácia </font>

V téme 2 sa budeme zaoberať metódami **aproximácie** (neznámych/zložitých) **funkcií** pomocou iných ("jednoduchších") funkcií (tzv. **interpolanty**). Častá je situácia, keď nejaká funkcia je daná len konečnou množinou bodov (hodnoty nezávislej premennej a odpovedajúce funkčné hodnoty) - napr. určité merania a pod., no v praxi potrebujeme určiť neznáme funkčné hodnoty v ďalších bodoch. 

**Polynomická interpolácia** (**aproximácia funkcií polynómami**) je starodávna prax, ale ťažké priemyselné použitie interpolácie sa začalo s **kubickými splajnami** v 20. storočí. Motivované praktikami **v lodiarskom a lietadlovom priemysle**, inžinieri Paul de Casteljau a Pierre Bézier u konkurenčných európskych výrobcoch automobilov Citroen a Renault, po ktorých nasledovali ďalší v General Motors v Spojených štátoch, podnietili vývoj toho, čo sa v súčasnosti nazýva **kubické splajny** a **Bézierové splajny**. 
<br/>
Aj keď boli vyvinuté pre aerodynamické štúdie automobilov, **splajny** sa používajú v mnohých aplikáciách vrátane **sádzania textu v počítači**. Revolúciu v tlači spôsobili dvaja inžinieri spoločnosti Xerox, ktorí založili spoločnosť s názvom Adobe a vydali jazyk PostScript™ v roku 1984. Dostalo sa to do pozornosti Steva Jobsa v spoločnosti Apple Corporation, ktorý hľadal spôsob, ako ovládať novo vynájdenú laserovú tlačiareň . **Bézierové splajny** boli jednoduchým nástrojom, ako prispôsobiť rovnaké **matematické krivky** písma viacerým rozlíšeniam tlačiarní. Neskôr spoločnosť Adobe použila mnoho základných myšlienok PostScriptu ako základ flexibilnejšieho formátu s názvom PDF (*Portable Document Format* - formát prenosného dokumentu), ktorý sa stal všadeprítomným typom súboru dokumentov do začiatku 21. storočia. 

 
$$\text{Zdroj: Sauer (2011).}$$

Systém globálneho polohovania (*Global Positioning System* - GPS) je **technológia určovania polohy** založená na satelitoch, ktorá poskytuje presné umiestnenie kedykoľvek, z ľubovoľného bodu na Zemi. Za niekoľko rokov prešla GPS z navigačnej technológie so špeciálnym účelom, ktoré používajú piloti, kapitáni lodí a turisti na každodenné použitie v automobiloch, mobilných telefónoch či PDA. <br/>
Systém pozostáva z 24 satelitov po presne regulovaných obežných dráhach, ktoré emitujú synchronizované signály. Prijímač na Zemi zachytáva satelitné signály, nájde svoju vzdialenosť od všetkých viditeľných satelitov a údaje používa na trianguláciu svojej polohy. K tomu, aby mohli GPS zariadenia spoľahlivo fungovať, sa využíva iný typ aproximácie funkcie - ide o tzv. **metódu najmenších štvorcov**. 

 

Tému 2 začneme **interpoláciou pomocou polynómov**. Rozoberieme si vlastnosti, výhody i nevýhody **Taylorovho polynómu**, **Lagrangeovho polynómu**, **Newtonovho** tvaru **interpolačného polynómu** a spomenieme i **barycentrickú formulu**. Budeme analyzovať **chybu polynomickej interpolácie** a ukážeme si, ako v niektorých prípadoch dosiahnuť lepšie výsledky implementovaním **Chebysevovych bodov**. <br/> 
Ďalším nástrojom ako vylepšiť nedostatky interpolačného polynómu sú tzv. **po častiach polynomické funkcie** alebo **splajny**. Detailnejšie si predstavíme **lineárne** a **kubické splajny**. <br/> 
Zatiaľ čo v doteraz uvedených metódach je dôležitou črtou to, že **interpolant** (resp. jeho graf) prechádza vopred danými bodmi (tzv. **interpolačné uzly**), tak existuje aj ďalší prístup ku aproximácii nejakej funkcie. Totiž funkcia, ktorá je aproximáciou inej funkcie, nemusí nutne prechádzať všetkými zadanými bodmi. Niekedy stačí, že je v istom zmysle blízko k týmto bodom. 
K nájdeniu takých aproximácií slúži napr. **metóda najmenších štvorcov** (MNŠ).
