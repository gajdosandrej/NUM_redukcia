**23. Dokážte nasledujúce tvrdenie súvisiace s chybou aproximácie pomocou interpolačného polynómu:** 

Nech $x_i=a+ih$ pre $i=0,1,\ldots,n$, pričom $h=(b-a)/n$. Potom pre ľubovoľné $x\in\langle a,b \rangle$ platí 

$$\displaystyle\prod\limits_{i=0}^{n}|x-x_i|\le\dfrac{1}{4}h^{n+1}n!.$$ 

Pomôcka - využite nasledujúce vzťahy: pre $x_j\le x \le x_{j+1}$ platí $|x-x_j||x-x_{j+1}|\le\dfrac{h^2}{4}$; ak $0\le j \le n-1$, potom $(j+1)!(n-j)!\le n!$.
