# <font color=brown> Motivácia </font>

Vykopaná tabuľka s klinovým písmom ukazuje, že Babylončania vypočítali odmocninu z čísla dva správne až na päž desatinných miest. V téme 5 predstavíme **numerické iteračné metódy**, ktoré môžu byť použité aj pre tento účel a v súčasných počítačoch sa stále využívajú. <br/> 
**Platforma Stewart**, robot so šiestimi stupňami voľnosti, ktorý môže byť lokalizovaný s extrémnou presnosťou, bola pôvodne vyvinutá Ericom Goughom zo spoločnosti Dunlop Tire Corporation v 50. rokoch minulého storočia na **testovanie pneumatík lietadiel**. Dnes sa jeho aplikácie pohybujú od **leteckých simulátorov**, ktoré sú často značne masové, až po **lekárske a chirurgické aplikácie**, kde je presnosť veľmi dôležitá. Riešenie problému **doprednej kinematiky** vyžaduje určenie polohy a orientácie plošiny vzhľadom na dĺžky vzpier. K tomu sú potrebné **presné metódy na hľadanie koreňov rovníc**. 

 

S potrebou dostatočne presne riešiť **(nelinárne) rovnice** sa môžeme stretnúť napr. aj **vo financiách** pri výpočte úrokových mier, **vo fyzikálnych modeloch** (stavová rovnica plynu) alebo tiež **v modeloch populačnej dynamiky** (Verhulstov logistický model) atď. 

 

V rámci témy 5 si priblížime tieto **iteračné metódy na riešenie nelineárnych rovníc: metóda bisekcie (polenia intervalu), Newtonova metóda (dotyčnicová), metóda sečníc, metóda regula falsi (tetivová) a metóda prostej iterácie (postupných aproximácií).** Preskúmame **rýchlosť konvergencie** uvedených metód a tiež to, za akých podmienok konvergujú k hľadanému koreňu rovnice. Nebudú chýbať ilustračné príklady konkrétnych výpočtov v Python-e.

### Základné 


* Rozumieť základným princípom vybraných iteračných metód pre riešenie nelineárnej rovnice. 


* Definovať rôzne typy rýchlostí konvergencie (číselných) postupností a rozumieť súvisu s konvergenciou iteračných metód. 


* Poznať vlastnosti vybraných iteračných metód vrátane rýchlosti konvergencie (za určitých podmienok). 


* Mať prehľad o tzv. zastavovacích kritériách iteračných metód a vedieť ich implementovať. 


* Vedieť odkrokovať (v Python-e) výpočet pomocou vybraných iteračných metód. 


* Použiť vhodný Python-ovský výpočtový balík na riešenie konkrétnej nelineárnej rovnice. 


### Pokročilé 


* Vedieť implementovať v Python-e vybrané iteračné metódy na riešenie nelineárnej rovnice. 


* Porovnať (aj graficky) presnosť resp. rýchlosť konvergencie rôznych iteračných metód (na konkrétnom príklade). 


* Odhadnúť počet iterácií konkrétnej metódy, potrebných na získanie aproximácie koreňa rovnice s danou presnosťou. 


* Pri metóde prostej iterácie vedieť overiť (ne)splnenie predpokladov pre existenciu pevného bodu a konvergenciu metódy. 


* Aplikovať iteračné metódy na vyriešenie úlohy s kontextom.
