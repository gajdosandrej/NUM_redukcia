# <font color=brown> Motivácia </font>

Pri **detekcii signálu** v bezdrôtových komunikačných systémoch s mnohými anténami a mnohými používateľmi sa vyžaduje riešenie systému lineárnych rovníc. Počet premenných sa môže pohybovať rádovo okolo hodnoty 2000. To sa nezdá byť dramaticky vysoké číslo, avšak výpočet musí byť veľmi rýchly.

 

**Iteračné metódy riešenia sústav lineárnych rovníc** sú aplikované pri (fyzikálnych) **animáciách deformácií**, **interakcií a pohybov objektov**, napr. vo videohrách. Výzvou je tu striktná požiadavka na čas výpočtu, ktorý musí byť pod niekoľko milisekúnd. V týchto prípadoch sú preferované lineárne iteračné metódy, pretože poskytujú približné riešenia v rámci danej tolerancie chýb a v krátkom čase.  

 

Ako sme naznačili, prax si mnohokrát vyžaduje efektívne riešenie veľkých systémov lineárnych rovníc, kde aplikácia priamych metód môže viesť k vysokým nárokom na počet operácií i pamäť. V takom prípade je vhodnejšie použiť **iteračné metódy** na riešenie príslušných sústav lineárnych rovníc, s ktorými sa oboznámime v téme 7. <br/>
Konkrétne sa budeme zaoberať týmito metódami: **Jacobiho, Gaussova-Seidelova, SOR či gradientné metódy - najväčšieho spádu i združených smerov**. Preskúmame tiež **konvergnečné vlastnosti** uvedených prístupov.
