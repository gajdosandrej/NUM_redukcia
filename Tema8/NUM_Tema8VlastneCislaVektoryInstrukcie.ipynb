# <font color=brown> Motivácia </font>

World Wide Web sprístupňuje bežnému používateľovi obrovské množstvo informácií – v skutočnosti také rozsiahle, že navigácia pomocou výkonného **vyhľadávacieho nástroja** je nevyhnutná. Technológia tiež poskytla miniaturizáciu a lacné senzory, vďaka čomu majú výskumníci k dispozícii veľké množstvo údajov. Ako možno efektívne využiť prístup k veľkému množstvu informácií? 

Mnoho aspektov **technológie vyhľadávania** a objavovania vedomostí vo všeobecnosti profituje z toho, že sa k nej pristupuje ako k problému **vlastných hodnôt** alebo **singulárných hodnôt**. Numerické metódy na vyriešenie týchto problémov s vysokou dimenziou generujú projekcie do podpriestorov nižšej dimenzie. Toto je presne to zjednodušenie, ktoré komplexné dátové prostredia najviac potrebujú. 

 

**Vlastné čísla** sa využívajú i v **komunikačných systémoch** na určenie teoretického limitu množstva informácií, ktoré možno preniesť cez komunikačné médium, ako je Vaša telefónna linka alebo vzduch. K tomu je nevyhnutný výpočet **vlastných vektorov** a **vlastných hodnôt** komunikačného kanála (reprezentovaného maticou). 

 

**Analýza hlavných komponentov** (ang. *Principal Component Analysis* - PCA) je jednou z kľúčových metód, ktorá sa používa na redukciu dimenzie dát, bez straty cenných informácií. Hlavná časť PCA je založená na koncepte výpočtu **vlastných vektorov a vlastných hodnôt kovariančnej matice atribútov**. <br/> 
S tým súvisí tiež aplikácia vlastných čísel v **zhlukovej analýze**, kedy je potrebné množinu objektov reprezentovanú atribútmi automaticky roztriediť do zmysluplných skupín/kategórií. 

V neposlednom rade sú vlastné a/alebo singulárne čísla matíc potrebné pri **kompresii obrázkov** či pri **rozpoznávaní tvárí**. 

 

Tému 8 začneme pripomenutím dôležitých poznatkov o vlastných číslach i vlastných vektoroch pre všeobecné matice ale tiež pre špeciálne, často používané matice - symetrické, pozitívne definitné matice. V rámci **lokalizácie vlastných čísel** uvedieme **Gerschgorinovu vetu**. Pozrieme sa na **podmienenosť problému výpočtu vlastných hodnôt**, s čím súvisí aj **Bauerova-Fikkova veta**. Rozoberieme **vplyv** niektorých **transformácií na vlastné čísla i vektory** a preskúmame tiež určovanie vlastných čísel vybraných špeciálnych matíc, kde to ide pomerne jednoducho. Ďalej sa budeme venovať konkrétnym **iteračným numerickým metódam na výpočet vlastných hodnôt** (prípadne aj vlastných vektorov). Pre výpočet jedného (napr. dominatného) vlastného čísla budú určené tieto metódy, ktoré sú priamymi modifikáciami tzv. **mocninnej** resp. **mocninovej metódy** - **metóda inverznej iterácie**, **metóda Rayleighovych podielov**. Za účelom výpočtu niekoľkých (viacerých) vlastných čísel matice budú k dispozícii tieto metódy: **simultánne** prípadne **ortogonálne iterácie**, **QR metóda**. V krátkosti sa budeme venovať i výpočtu **singulárnych hodnôť** a **singulárnych vektorov**.

### Základné 


* Poznať základné pojmy súvisiace s vlastnými čislami a vektormi 


* Poznať základné transformácie matíc a vedieť ako sa ich aplikovaním zmenia vlastné hodnoty či vlastné vektory matice.  


* Určiť vlastné čísla špeciálnych matíc (diagonálne a trojuholníkové matice, Schurov rozklad matice, blokovo-trojuholníkové matice). 


* Rozumieť princípom vybraných numerických metód pre výpočet jedného (dominantného) vlastného čísla a príslušného vlastného vektora matice. 


* Rozumieť princípom vybraných numerických metód pre výpočet viacerých (všetkých) vlastných čísel a príslušných vlastných vektorov matice. 


* Poznať výhody/nevýhody vybraných algoritmov na približný výpočet vlastných hodnôt matice i príslušných vlastných vektorov. 

### Pokročilé 


* Vysloviť a aplikovať Gerschgorinovu vetu na lokalizáciu vlastných hodnôt matice (prípadne využiť na určenie vhodného posunu pri modifikácii mocninovej metódy). 


* Určiť podmienenosť vlastných čísel i vektorov (dôkaz i aplikovanie Bauerovej-Fikovej vety, číslo podmienenosti vlastného čísla matice). 


* Vedieť implementovať vybrané numerické metódy pre výpočet jedného (dominantného) vlastného čísla a príslušného vlastného vektora matice.

* Naprogramovať QR rozklad matice. 


* Rozumieť singulárnemu rozkladu matice a vedieť ho naprogramovať. 


* Aplikovať vhodné metódy na riešenie jednoduchých úloh s kontextom, ktoré súvisia s výpočtom vlastných čísel a vlastných vektorov.
