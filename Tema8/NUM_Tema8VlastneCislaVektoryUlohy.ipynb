**25. (Kompresia obrázka).** Problému kompresie obrazu možno čeliť použitím singulárneho rozkladu matice. Čiernobiely obrázok sa dá reprezentovať obdĺžnikovou maticou $\mathbf{A}$ s rozmermi $m\times n$ , kde $m$ a $n$ predstavujú počet pixelov, ktoré sú prítomné v horizontálnom i vertikálnom smere a koeficient $a_ij$ predstavuje intenzitu šedej $(i,j)$-tého pixelu. Ak vezmeme do úvahy singulárny rozklad matice $\mathbf{A}=\mathbf{U}\mathbf{\Sigma}\mathbf{V}'$ a ak označíme $\mathbf{u}_i$, $\mathbf{v}_i$ $i$-té stĺpcové vektory matíc $\mathbf{U}$, $\mathbf{V}$, tak dostaneme

\begin{align*}
\mathbf{A}=\sigma_1\mathbf{u}_1\mathbf{v}_1'+\sigma_2\mathbf{u}_2\mathbf{v}_2'+\ldots+\sigma_p\mathbf{u}_p\mathbf{v}_p'.
\end{align*}

Maticu $\mathbf{A}$ je možne aproximovať maticou $\mathbf{A}_k$, ktorá vznikne skrátením súčtu uvedeného vo vzťahu vyššie, prčom sa vezme do úvahy prvých $k$ členov, kde $1\le k \le p$. Ak sú singulárne čísla $\sigma_i$ zoradené v poradí $\sigma_1\ge\sigma_2\ge\ldots\ge\sigma_p$, pričom sa zanedbá zvyšných (posledných) $p-k$ singulárnych čísel, tak by toto skrátenie/vynechanie nemalo výrazne ovplyvniť kvalitu obrázka. Prakticky to znamená, že na prenos (napr. z jedného počítača do iného) komprimovaného obrázka reprezentovaného maticou $\mathbf{A}_k$, stačí preniesť vektory $\mathbf{u}_i$, $\mathbf{v}_i$ a singulárne čísla $\sigma_i$ pre $1\le i \le k$.
